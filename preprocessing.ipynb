```python
data_path = "/content/TimKoornstra_synthetic-financial-tweets-sentiment_train.csv"
df = pd.read_csv(data_path)
```

```python
df = df.rename(columns={"tweet": "text", "sentiment": "label"})
df.head()
```

```python
df['text'] = df['text'].apply(emoji.demojize)
df['text']
```

```python
df['lowercase_text'] = df['text'].str.lower()
df[['text', 'lowercase_text']]
```

```python
df['no_punctuation'] = df['lowercase_text'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
df[['lowercase_text', 'no_punctuation']]
```

```python
df['no_num'] = df['no_punctuation'].apply(lambda x: re.sub(r'\d+', '', x))
df[['no_punctuation', 'no_num']]
```
